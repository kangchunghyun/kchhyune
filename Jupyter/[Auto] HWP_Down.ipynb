```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
from selenium.common.exceptions import NoSuchElementException
import time
```

```python
# Set up Chrome options (optional)
chrome_options = Options()
chrome_options.add_argument("--start-maximized")  # Opens the browser in maximized mode

# Initialize the WebDriver
driver = webdriver.Chrome(options=chrome_options)  # No need for Service object explicitly
```

```python
# Define the URL you want to visit
url = "https://job.alio.go.kr/recruitview.do?pageNo=2&idx="
idx = 280501
driver.get(url + str(idx))

while True:

    # Pause to allow the page to load
    time.sleep(3)

    attach_button = driver.find_element(By.XPATH, '//*[@id="contentTab"]/ul/li[3]')
    attach_button.click()

    try: 
        for num in range(1, 5):
            down_button = driver.find_element(By.XPATH, f'//*[@id="tab-3"]/table/tbody/tr[{num}]/td/p/a')
            down_button.click()
            time.sleep(3)

    except NoSuchElementException as nofile:
        print(f"첨부된 파일이 없습니다." )

    idx = idx - 1
    driver.get(url + str(idx))

    # Pause to allow the page to load
    time.sleep(3)
```

```python
try: 
    for num in range(1, 5):
        down_button = driver.find_element(By.XPATH, f'//*[@id="tab-3"]/table/tbody/tr[{num}]/td/p/a')
        down_button.click()
        time.sleep(3)
        
except NoSuchElementException as nofile:
    print(f"첨부된 파일이 없습니다." )
```
