```python
import requests
from json import loads
```

```python
url = "https://mb-api.abuse.ch/api/v1/"
downDir = input("다운로드 받을 경로를 입력하세요: ")
```

```python
tag = input("다운로드 받을 확장자를 입력하세요: ")
```

```python
sha256_hash_tag = [] # 다운로드 받을 sha256 목록
sha256_hash_fileType = []

get_taginfo = {

    "apikey": "cf95fc11f48630ab31d27f81390175fc",
    #"first_seen": "2021-01-01 00:00:00",
    #"last_seen": "2022-01-01 00:00:00",
    "query": "get_taginfo", # get_taginfo = 리스트 조회, get_file = 다운로드
    "tag": "CVE-2017-11882", # 조회할 확장자 설정
    #"file_type": "docx",
    "limit": 1000,
}


get_file_type = {
    
    "apikey": "cf95fc11f48630ab31d27f81390175fc",
    "query": "get_file_type",
    "file_type": tag,
    "limit": 1000
    
}
```

```python
response = requests.post(url, data=get_taginfo).json()

for i in range(len(response["data"])):
    sha256_hash_tag.append(response["data"][i]["sha256_hash"])
    
print(len(sha256_hash_tag))
```

```python
response_fileType = requests.post(url, data=get_file_type).json()

for i in range(len(response_fileType["data"])):
    sha256_hash_fileType.append(response_fileType["data"][i]["sha256_hash"])
    
print(len(sha256_hash_fileType))
```

```python
down_bot = open(downDir + f"LIST_{tag}.bat", 'w') 

count = 0

#for i in sha256_hash_fileType: 
for i in sha256_hash_tag:
    
    count = count + 1
    wget = f'wget --post-data "query=get_file&sha256_hash={i}" -O "{i}.zip" https://mb-api.abuse.ch/api/v1/'
    print("( "+ str(count) + " )" + wget)
    down_bot.write(wget+"\n")
    
    
down_bot.close()
```
